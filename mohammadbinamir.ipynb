```python
def find_max(*numbers):
    if not numbers:
        return None
    return max(numbers)
print(find_max(5, 12, 87, 3, 42))  
```

```python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(3))
```

```python
def apply_function(func, numbers):
    result = []
    for num in numbers:
        result.append(func(num))
    return result

def square(x):
    return x * x

my_nums = [1, 2, 3, 4]
print(apply_function(square, my_nums)) 
```
